```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

def main():
    # Generate synthetic data
    np.random.seed(42)
    X = np.random.rand(100, 2) * 10  # 100 samples, 2 features
    y = (X[:, 0] + X[:, 1] > 10).astype(int)  # Binary target based on a simple rule

    # Standardize the features
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    # Plot the data
    plt.figure(figsize=(8, 6))
    plt.scatter(X_scaled[y == 0, 0], X_scaled[y == 0, 1], label='Class 0', alpha=0.7)
    plt.scatter(X_scaled[y == 1, 0], X_scaled[y == 1, 1], label='Class 1', alpha=0.7)
    plt.title('Synthetic Data Visualization')
    plt.xlabel('Feature 1 (Standardized)')
    plt.ylabel('Feature 2 (Standardized)')
    plt.legend()
    plt.grid()
    plt.show()

if __name__ == "__main__":
    main()
```
